```python
import warnings
import os
warnings.filterwarnings('ignore', category=FutureWarning)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
from twilbert.utils.generator import PretrainingGenerator
from twilbert.utils.utils import Utils
from twilbert.models.twilbert_model import TWilBertModel
from twilbert.preprocessing.tokenizer import (FullTokenizer,
                                              convert_ids_to_tokens)
from twilbert.utils.pretraining_metrics import acc_m
from matplotlib import pyplot as plt
from keras import backend as K
import matplotlib.gridspec as gridspec
import numpy as np
import seaborn as sns
from collections import Counter
import json
import sys
from IPython.display import clear_output
%matplotlib inline
```

```python
# Dataset #
dataset_file = "./pretraining_corpora/samples.tsv"
vocab_file = "./vocab"

# Representation #
max_len = 128
max_len_test = 128
tokenizer = FullTokenizer(vocab_file)
vocab_size = len(tokenizer.vocab)
bucket_min_a = 5
bucket_min_b = 5
bucket_max_a = 45
bucket_max_b = 45
bucket_steps = 5

# Model #
factorize_embeddings = False
cross_sharing = False
embedding_size = 768
hidden_size = 768
n_encoders = 6
n_heads = 6
attention_size = 64
input_dropout = 0.0
output_dropout = 0.0

pkm = False
pkm_params = {}

mlm_type = "span"
mlm_max_span = 3
mlm_budget = 0.15
mlm_probs = [0.8, 0.1, 0.1]

rop_n_hidden = 0
rop_hidden_size = -1

path_load_weights = "./pretrained_weights/model_04-2.4318-0.5865.hdf5"

encoder_size = [hidden_size for i in range(n_encoders)]
attention_size = [attention_size for i in range(n_encoders)]
n_heads = [n_heads for i in range(n_encoders)]
```

```python
# Model definition #
twilbert_model = TWilBertModel(max_len, vocab_size, embedding_size,
                               encoder_size, attention_size,
                               n_heads, cross_sharing,
                               factorize_embeddings,
                               input_dropout, output_dropout,
                               rop_n_hidden, rop_hidden_size,
                               None, None, pkm, pkm_params,
                               input_length=None)

twilbert_model.build()
model = twilbert_model.model

twilbert_model.compile(model)

twilbert_model.load(model, path_load_weights)

print(model.summary())
```

```python
da = PretrainingGenerator(dataset_file, tokenizer,
                          256, mlm_type,
                          mlm_max_span, mlm_budget,
                          mlm_probs, bucket_min_a,
                          bucket_min_b, bucket_max_a,
                          bucket_max_b, bucket_steps)
```

### La idea es hacer una "interpolación" entre una distribución uniforme y la distribución que predice el generador

```python
c = np.random.randint(0, 255)
gen = da.generator()
(batch_x, position_indices, segment_indices), \
    (batch_rop, batch_mlm) = next(gen)
x = batch_x[c]
y_rop = batch_rop[c]
ref = batch_mlm[c].squeeze(axis=-1)
```

```python
preds = model.predict([np.array([x]),
                       np.array([position_indices[0]]),
                       np.array([segment_indices[0]])])
pred_rop = preds[0]
pred_mlm = preds[1][0]
```

```python
# TWilBERT distribution for the first masked token of a random sample #
masked_index = np.where(ref != 0)[0][0]
y_gen = pred_mlm[masked_index]
x_gen = np.arange(0, len(y_gen))

# Uniform distribution 
y_uni = np.ones(len(y_gen)) / len(y_gen)
x_uni = np.arange(0, len(y_uni))
```

```python
# The idea is to approximate a "dumb" distribution (uniform) to the "smart" distribution
# (computed by the generator after pre-training), in order to simulate the training process of the generator.
# The confidence of the generator improves with each training step during pre-training, so we have to simulate it during the approximation.
# This confidence can be viewed as increments if the probability of a token in the approximation have to be increased to reach the target probability,
# or as reductions if the probability of a token in the approximation have to be reduced.
# It is required to tune the speed of the approximations. This allows to sample more dumb tokens during the 
# approximation until reaching the target distribution.

# A naive solution consists in making the approximations slower when higher increments or reductions have to be made:

# 1) If the probability has to be increased: slower growth when more probability has the token in the target distribution.
# 2) If the probability has to be reduced: slower growth when less probability has the token in the target distribution.

# Be p_trg the target distribution, p_u the uniform distribution, p_t the distribution computed at timestep t,
# lr the hyper-parameter to control the speed of the approximation:

# p_0 = p_u
# p_t = (p_t - p_trg) * lr

# The disadvantage of this is that lr is a global hyper-parameter, and all the tokens are approximated with the same speed.
# To solve this, we propose to split lr in two factors: alpha, the general speed of the process and w the weight
# that controls the relative speed (greater when higher or lower is the probability in the target distribution).

# ... #

p_dest = y_gen
p_ini = y_uni
alpha = 1e-4
epsilon = 3e-2
samples_process = []
samples_target = [] # Visualization only
print("Inicial:", sum((p_dest - p_ini)))


# Previo al proceso iterativo, se sabe cuales decrecen y cuales crecen. También se saben los máximos incrementos
# y máximas reducciones.

# Calcular diferencia
diff = p_dest - p_ini

# Ver cuáles suben y bajan
lr = np.sign(diff)
ups = np.where(lr==1)
downs = np.where(lr==-1)

# Calcular máximas subidas y reducciones y mínimas subidas y reducciones para normalizar
max_up = np.max(p_dest[ups])
max_down = np.max(p_dest[downs])
min_up = np.min(p_dest[ups])
min_down = np.min(p_dest[downs])

# El learning rate se puede calcular desde el inicio
lr[ups] =  alpha * (np.abs(1. - ((p_dest[ups] - min_up) / (max_up  + epsilon - min_down))))
lr[downs] = alpha * ((p_dest[downs] - min_down) / (max_down + epsilon - min_down))

for i in range(1000000):
    
    # Aproximar p_t a p_target
    p_ini += (p_dest - p_ini) * lr
    
    # Normalizar p_t
    p_ini = p_ini / p_ini.sum()
    
    samples_process.append(np.random.choice(x_gen, p = p_ini))
    samples_target.append(np.random.choice(x_gen, p = p_dest))
    
    if i % 5000 == 0:
        gs = gridspec.GridSpec(3, 2)
        fig = plt.figure(figsize=(20, 8))
        ax1 = plt.subplot(gs[0, :])
        ax2 = plt.subplot(gs[1, :])
        ax3 = plt.subplot(gs[2, :])

        # Dibujar la aproximación #
        
        ax1.plot(x_gen, y_gen)
        ax1.plot(x_uni, p_ini)
        ax1.set_title("Distributions at it %d" % i)
        ax1.set_xlim(0, len(x_gen))
    
        # Dibujar el histograma de la distribucion hasta t #
        
        count_proc = Counter(samples_process)
        proc_vals = [0 for j in range(len(x_gen))]
        for j in count_proc:
            proc_vals[j] = count_proc[j]
        ax2.set_title("Sampling Histogram p_t at it %d" % i)
        ax2.set_xlim(0, len(x_gen))
        ax2.plot(np.arange(len(x_gen)), proc_vals)
        
        # Dibujar el histograma de la distribución destino #
        
        count_target = Counter(samples_target)
        target_vals = [0 for j in range(len(x_gen))]
        for j in count_target:
            target_vals[j] = count_target[j]
        ax3.set_title("Sampling Histogram p_trg at it %d" % i)
        ax3.set_xlim(0, len(x_gen))
        ax3.plot(np.arange(len(x_gen)), target_vals)
        clear_output(wait=True)
        
        fig.tight_layout(pad=3.0)
        plt.show()
```

```python
def temp_softmax(preds, temperature=1.0):
    # helper function to sample an index from a probability array
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    probs = exp_preds / np.sum(exp_preds)
    return probs

def compute_temperature(t, temp, max_steps):
    if t <= max_steps:
        return 1 + (1./2.) * (1. + np.cos((t * np.pi) / max_steps)) * (T-1)
    return 1.

c = np.random.randint(0, 255)
gen = da.generator()
(batch_x, position_indices, segment_indices), \
    (batch_rop, batch_mlm) = next(gen)
x = batch_x[c]
y_rop = batch_rop[c]
ref = batch_mlm[c].squeeze(axis=-1)

preds = model.predict([np.array([x]),
                       np.array([position_indices[0]]),
                       np.array([segment_indices[0]])])
pred_rop = preds[0]
pred_mlm = preds[1][0]

# TWilBERT distribution for the first masked token of a random sample #
masked_index = np.where(ref != 0)[0][0]
y_gen = pred_mlm[masked_index]
x_gen = np.arange(0, len(y_gen))

# Uniform distribution 
y_uni = np.ones(len(y_gen)) / len(y_gen)
x_uni = np.arange(0, len(y_uni))
# The idea is to approximate a "dumb" distribution (uniform) to the "smart" distribution
# (computed by the generator after pre-training), in order to simulate the training process of the generator.
# The confidence of the generator improves with each training step during pre-training, so we have to simulate it during the approximation.
# This confidence can be viewed as increments if the probability of a token in the approximation have to be increased to reach the target probability,
# or as reductions if the probability of a token in the approximation have to be reduced.
# It is required to tune the speed of the approximations. This allows to sample more dumb tokens during the 
# approximation until reaching the target distribution.

# A naive solution consists in making the approximations slower when higher increments or reductions have to be made:

# 1) If the probability has to be increased: slower growth when more probability has the token in the target distribution.
# 2) If the probability has to be reduced: slower growth when less probability has the token in the target distribution.

# Be p_trg the target distribution, p_u the uniform distribution, p_t the distribution computed at timestep t,
# lr the hyper-parameter to control the speed of the approximation:

# p_0 = p_u
# p_t = (p_t - p_trg) * lr

# The disadvantage of this is that lr is a global hyper-parameter, and all the tokens are approximated with the same speed.
# To solve this, we propose to split lr in two factors: alpha, the general speed of the process and w the weight
# that controls the relative speed (greater when higher or lower is the probability in the target distribution).

# ... #

p_dest = y_gen
p_ini = y_uni
alpha = 1e-4
epsilon = 3e-2
print("Inicial:", sum((p_dest - p_ini)))
samples_process = []
samples_target = [] # Visualization only
samples_target_temperature = []
max_steps_temperature = 20000
T = 5
# Previo al proceso iterativo, se sabe cuales decrecen y cuales crecen. También se saben los máximos incrementos
# y máximas reducciones.

# Calcular diferencia
diff = p_dest - p_ini

# Ver cuáles suben y bajan
lr = np.sign(diff)
ups = np.where(lr==1)
downs = np.where(lr==-1)

# Calcular máximas subidas y reducciones y mínimas subidas y reducciones para normalizar
max_up = np.max(p_dest[ups])
max_down = np.max(p_dest[downs])
min_up = np.min(p_dest[ups])
min_down = np.min(p_dest[downs])

# El learning rate se puede calcular desde el inicio
lr[ups] =  alpha * (np.abs(1. - ((p_dest[ups] - min_up) / (max_up  + epsilon - min_down))))
lr[downs] = alpha * ((p_dest[downs] - min_down) / (max_down + epsilon - min_down))

# Aproximar p_t a p_target en el timestep t #
for t in range(0, 1500000, 100):
    p_ini = p_ini + (np.clip(lr * t, a_min=0, a_max=1)) * (p_dest - p_ini)

    # Normalizar p_t
    #p_ini = p_ini / p_ini.sum()
    temp = compute_temperature(t, T, max_steps_temperature)
    for _ in range(100):
        samples_process.append(np.random.choice(x_gen, p = p_ini / p_ini.sum()))
        samples_target.append(np.random.choice(x_gen, p = p_dest))
        samples_target_temperature.append(np.random.multinomial(1, temp_softmax(p_dest, temp), 1)[0].argmax())

    
    gs = gridspec.GridSpec(4, 2)
    fig = plt.figure(figsize=(20, 8))
    ax1 = plt.subplot(gs[0, :])
    ax2 = plt.subplot(gs[1, :])
    ax3 = plt.subplot(gs[2, :])
    ax4 = plt.subplot(gs[3, :])
    
    # Dibujar la aproximación #

    ax1.plot(x_gen, y_gen)
    ax1.plot(x_uni, p_ini)
    ax1.set_title("Distributions at it %d" % t)
    ax1.set_xlim(0, len(x_gen))

    # Dibujar el histograma de la distribucion hasta t #

    count_proc = Counter(samples_process)
    proc_vals = [0 for j in range(len(x_gen))]
    for j in count_proc:
        proc_vals[j] = count_proc[j]
    ax2.set_title("Sampling Histogram p_t at it %d" % t)
    ax2.set_xlim(0, len(x_gen))
    ax2.plot(np.arange(len(x_gen)), proc_vals)

    # Dibujar el histograma de la distribución destino #

    count_target = Counter(samples_target)
    target_vals = [0 for j in range(len(x_gen))]
    for j in count_target:
        target_vals[j] = count_target[j]
    ax3.set_title("Sampling Histogram p_trg at it %d" % t)
    ax3.set_xlim(0, len(x_gen))
    ax3.plot(np.arange(len(x_gen)), target_vals)
    

    # Dibujar el histograma de la distribución destino con temp softmax#

    count_target = Counter(samples_target_temperature)
    target_vals = [0 for j in range(len(x_gen))]
    for j in count_target:
        target_vals[j] = count_target[j]
    ax4.set_title("Sampling Histogram p_trg with temperature at it %d" % t)
    ax4.set_xlim(0, len(x_gen))
    ax4.plot(np.arange(len(x_gen)), target_vals)
    clear_output(wait=True)
    
    fig.tight_layout(pad=3.0)
    plt.show()
    
    
    
    """
    gs = gridspec.GridSpec(1, 1)
    fig = plt.figure(figsize=(20, 8))
    ax1 = plt.subplot(gs[0, :])

    # Dibujar la aproximación #  
    ax1.plot(x_gen, y_gen)
    ax1.plot(x_uni, p_ini)
    ax1.set_title("Distributions at it %d" % t)
    ax1.set_xlim(0, len(x_gen))
    clear_output(wait=True)
    plt.show()
    """
```
